Even though a circle is the limit process of polygon DLAs, it's actually the easiest to pick a point along it than it is for polygons.

I will pick a point along the circle and shoot particles toward the center. I have not decided how to store the data. Is this a discrete lattice or a real field? In the real case, I would want to store points in some sort of quad-tree for collision detection.

```python
from __future__ import division
import numpy as np
import numpy.random
from IPython.core.display import display, Math, Image
import scipy
import scipy.misc
```

```python
def incrange(start, stop):
    # Count forwards?
    if start <= stop:
        return xrange(start, stop+1)
    # Count backwards.
    return xrange(start, stop-1, -1)
```

```python
def neighbors(canvas, x, y):
    output = []
    for xi in incrange(x-1, x+1):
        for yi in incrange(y-1, y+1):
            if 0 <= xi < canvas.shape[0] and 0 <= yi < canvas.shape[1]:
                output.append(canvas[xi, yi])
    return output
```

```python
initialpoints = np.random.rand(250000)
```

```python
SIZE=800
WIDTH = SIZE*2+1
canvas = np.zeros((WIDTH, WIDTH))
centerx = SIZE
centery = SIZE
canvas[centerx, centery] = 1
canvas.shape
```

```python
initialcirclepoints = np.array(
    np.vstack(
        (np.floor(np.cos(2*np.pi*initialpoints) * SIZE + centerx),
         np.floor(np.sin(2*np.pi*initialpoints) * SIZE + centery))),
    dtype=np.int)
initialcirclepoints.T.shape
```

```python
# Take initial points and see where they hit the shape.
# I will use Bresenham's line algorithm to find this.
# http://en.wikipedia.org/wiki/Bresenham%27s_line_algorithm
# If any point has a neighbor in the shape already, we stop.
for x0, y0 in initialcirclepoints.T:
    #print "start", x0, y0
    x1 = centerx
    y1 = centery
    deltax = x1 - x0
    deltay = y1 - y0
    error = 0
    #print deltaerr
    linepoints = []
    if abs(deltax) >= abs(deltay):
        y = y0
        sy = 1 if y0 < y1 else -1
        # Assume deltax != 0 (line is not vertical),
        # note that this division needs to be done in a way that preserves the fractional part
        # Thus, from __future__ import division.
        deltaerr = abs(deltay / deltax)
        for x in incrange(x0, x1):
            linepoints.append((x,y))
            # Time to shift to a new line?
            error = error + deltaerr
            if error >= 0.5:
                y = y + sy
                error = error - 1.0
    else:
        x = x0
        sx = 1 if x0 < x1 else -1
        # Assume deltay != 0 (line is not horizontal),
        # note that this division needs to be done in a way that preserves the fractional part
        # Thus, from __future__ import division.
        deltaerr = abs(deltax / deltay)
        for y in incrange(y0, y1):
            linepoints.append((x,y))
            # Time to shift to a new line?
            error = error + deltaerr
            if error >= 0.5:
                x = x + sx
                error = error - 1.0
    
    for x, y in linepoints:
        #if x >= WIDTH or y >= WIDTH:
        #    print "start", x0, y0
        #    print "now",x, y
        #    assert "Tried to draw outside canvas!"
        # plot(x,y)
        #canvas[x, y] = 1.0
        # Is there a neighbor point? If so, add this point to the shape
        # and stop.
        if any(neighbors(canvas, x, y)):
            canvas[x, y] = 1.0
            break # Very important! This is what makes it DLA.
```

```python
scipy.misc.imsave("circle.png", canvas)
display(Image(filename="circle.png"))
```
